```python
import pandas as pd
import plotly.express as px
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/ratimayy/tools/main/car.csv')
df.head()
```

```python
a = df[(df['Brand']=='Ford') | (df['Brand']=='Toyota') | (df['Brand']=='Chevrolet') | (df['Brand']=='Honda') | (df['Brand']=='Nissan')]
a
```

```python
a.drop('Q1 2024', axis = 1, inplace = True)
a
```

```python
b = a.copy()
b.info()
b["Q2 2021"].replace(",","",inplace=True, axis = 0)
b
```

```python
import plotly.express as px

# Melt the DataFrame to convert it to long format
df_long = a.melt(id_vars=["Brand"], var_name="Quarter", value_name="Sales")

# Plot the bar chart using Plotly Express
fig = px.bar(df_long, x="Brand", y="Sales", color="Quarter",
             title="Car Sales by Brand and Quarter",
             labels={"Sales": "Number of Sales", "Brand": "Car Brand", "Quarter": "Quarter"})

# Set the Y-axis range
#fig.update_yaxes(range=[0, 500000])

# Show the plot
fig.show()

```

```python
Q_counts = df_long.groupby('Quarter', 'Sales']).size().reset_index(name='count')

fig = px.bar(Q_counts,
             x='Brand',
             y='Sales',
             color='Quarter')

fig.show()
```
